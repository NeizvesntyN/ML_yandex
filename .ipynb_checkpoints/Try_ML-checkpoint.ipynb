```python
df = pd.read_csv("datasets_eat_places.csv")

df.info()
```

```python
df = df.fillna(0)
new_df = df.drop(df[(df.average_bill > 2500) | (df.average_bill == 0)].index)
new_df.reset_index(drop=True, inplace=True)
new_df.info()
```

```python
# 3)
```

```python
# 1

print(new_df.city.value_counts())
print()
print(new_df.rubric.unique())
```

```python
# 2

new_df['rubric'].value_counts() # 5530
```

```python
rb = new_df.query('rubric == "Ресторан" or rubric == "Бар, паб"') \
    .groupby('rubric', as_index=False) \
    .agg({'average_bill': 'sum'}) 

int(rb.average_bill.sum() / 5530)
```

```python
# 3

all_ctg = 139
rl_ctg = int(new_df.query('rubric == "Кофе с собой"') \
    .agg({'coffee_to_go': 'sum'}) \
    .coffee_to_go)
f"{round(rl_ctg / all_ctg * 100)}%"
```

```python
#  4

avb = (
    new_df
    .groupby('city', as_index=False)
    .agg({'average_bill': 'sum'})
)
avb
# 4368000
```
